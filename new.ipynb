```python
import pandas as pd
```

```python
df = pd.read_excel("coffee (1).xlsx")
df.head()
```

```python
df.info()
```

```python
df["year"] = df["date"].dt.year
df.head()
```

```python
df["year"] = df["date"].dt.year
df.head()
```

```python
df["day_name"] = df["date"].dt.day_name()
df.head()
```

```python
def time_check(time):
    if time.hour < 12:
        return "Morning"
    elif time.hour < 16:
        return "Afternoon"
    else:
        return "Evening"
```

```python
df["time_of_day"] = df["datetime"].apply(time_check)
df.head()
```

```python
# HOW TO DROP?SELECT COLUMNS
# TO SELECT MULTIPLE COLUMNS
df = df[["month", "year", "day"]]
df.head()
```

```python
# TO DROP COLUMNS YOU DON'T WANT
df = df.drop(["date", "datetime", "card"], axis=1)
df.head()
```
